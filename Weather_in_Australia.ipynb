Зобразіть динаміку змін середньої температури в Південній Австралії (стовпець

```python
years = data['Year']
southAVt = data['southAVt']

max_value = southAVt.max()
min_value = southAVt.min()
max_year = data[data['southAVt'] == max_value]['Year'].values[0]
min_year = data[data['southAVt'] == min_value]['Year'].values[0]

sns.set_theme(style="whitegrid")
plt.figure(figsize=(15, 10))
sns.lineplot(x=years, y=southAVt, marker='o', linestyle='-', color='#FFB6C1', label='Southern temperature')

plt.title('Dynamics of Average Temperature Changes in Southern Australia', color='b')
plt.xlabel('Year', color='b')
plt.ylabel('Southern Region Average Temperature (degrees C)', color='b')
plt.scatter([max_year], [max_value], color='#FF6961', zorder=3, label='Max Temperature')
plt.scatter([min_year], [min_value], color='#AEC6CF', zorder=3, label='Min Temperature')
plt.legend(loc='upper center')
plt.show()
```

Як впливало кількість сонячних плям на середньорічну температуру в західній та
східній частинах Австралії? Придумайте підходящу візуалізацію.

```python
sunspot = data['sunspot']
west_temp = data['westAVt']
east_temp = data['eastAVt']


fig, axs = plt.subplots(1, 2, figsize=(18, 8))


sns.scatterplot(ax=axs[0], x=sunspot, y=west_temp, color='#C3B1E1', alpha=0.6)
sns.regplot(ax=axs[0], x=sunspot, y=west_temp, scatter=False, color='#C3B1E1')
axs[0].set_title('Influence of the number of sunspots on the average annual temperature (Western Australia)', fontsize=12, color='b')
axs[0].set_xlabel('Number of sun spots', fontsize=12, color='b')
axs[0].set_ylabel('Average annual temperature (West AVt)', fontsize=12, color='b')
axs[0].grid(True)


sns.scatterplot(ax=axs[1], x=sunspot, y=east_temp, color='#FFB6C1', alpha=0.6)
sns.regplot(ax=axs[1], x=sunspot, y=east_temp, scatter=False, color='#FFB6C1')
axs[1].set_title('Influence of the number of sunspots on the average annual temperature (Eastern Australia)', fontsize=12, color='b')
axs[1].set_xlabel('Number of sun spots', fontsize=12, color='b')
axs[1].set_ylabel('Average annual temperature (East AVt)', fontsize=12, color='b')
axs[1].grid(True)


plt.tight_layout()
plt.show()
```

Як розподілялося кількість опадів по регіонам Австралії в період з 1990 по 2012 рік?
Відповідним типом візуалізації буде графік-пиріг (див. plt.pie()). Також необхідні
деякі обчислення в масиві.

```python
data_period = data[(data['Year'] >= 1990) & (data['Year'] <= 2012)]

total_rainfall_west = data_period['westAVt'].sum()
total_rainfall_east = data_period['eastAVt'].sum()
total_rainfall_north = data_period['northAVt'].sum()
total_rainfall_south = data_period['southAVt'].sum()

total_rainfall = [total_rainfall_west, total_rainfall_east, total_rainfall_north, total_rainfall_south]

regions = ['West', 'East', 'North', 'South']

plt.figure(figsize=(10, 10))
plt.pie(total_rainfall, labels=regions, autopct='%1.1f%%', startangle=140, colors=['#C3B1E1', '#FDFD96', '#FFB6C1', '#FF6961'])
plt.title('Distribution of Rainfall Across Regions of Australia (1990-2012)', fontsize=16, color = 'b')
plt.axis('equal')
plt.show()
```

На цьому графіку зображено відсотковий розподіл опадів в усіх 4 частинах Австралії.
Найбільше опадів на півночі, а найменше на півдні.

У якому з регіонів кількість опадів та середня температура корелюють більше
всього? Придумайте підходящу візуалізацію.

```python
corr_rainfall_west = data['westAVt'].corr(data['westRain'])
corr_rainfall_east = data['eastAVt'].corr(data['eastRain'])
corr_rainfall_north = data['northAVt'].corr(data['northRain'])
corr_rainfall_south = data['southAVt'].corr(data['southRain'])

correlation_dict = {
    'West': corr_rainfall_west,
    'East': corr_rainfall_east,
    'North': corr_rainfall_north,
    'South': corr_rainfall_south
}

max_correlation_region = max(correlation_dict, key=correlation_dict.get)

print(correlation_dict)
```

```python
regions = ['westAVt', 'eastAVt', 'northAVt', 'southAVt']
rainfall = ['westRain', 'eastRain', 'northRain', 'southRain']
colors = ['#C3B1E1', '#77DD77', '#FFB6C1', '#FF6961']

bubble_data = pd.DataFrame({
    'Temperature': pd.concat([data[region] for region in regions], ignore_index=True),
    'Rainfall': pd.concat([data[rain] for rain in rainfall], ignore_index=True),
    'Region': ['West'] * len(data) + ['East'] * len(data) + ['North'] * len(data) + ['South'] * len(data)
})

color_map = {'West': '#C3B1E1', 'East': '#77DD77', 'North': '#FFB6C1', 'South': '#FF6961'}
bubble_data['Color'] = bubble_data['Region'].map(color_map)

plt.figure(figsize=(10, 6))
for region, color in color_map.items():
    subset = bubble_data[bubble_data['Region'] == region]
    plt.scatter(subset['Temperature'], subset['Rainfall'],
                s=subset['Rainfall']*2,
                c=color,
                label=region,
                alpha=0.5,
                edgecolors='w',
                linewidth=0.5)

plt.xlabel('Temperature (°C)')
plt.ylabel('Rainfall (mm)')
plt.title('Bubble Chart of Rainfall vs Temperature Across Regions')
plt.legend(title='Region')
plt.grid(True)
plt.show()
```
