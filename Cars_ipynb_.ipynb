```python
import seaborn as sns
import matplotlib.pyplot as plt

brand_counts = data['brand'].value_counts()
top_10_brands = brand_counts.head(10)
print('brand counts')
print(brand_counts)
print('top 10 brands')
print(top_10_brands)
```

```python
brands_unique = brand_counts[brand_counts == 1]
print(brands_unique)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_brands.values, y=top_10_brands.index, palette='viridis')
for index, value in enumerate(top_10_brands.values):
    plt.text(value, index, str(value), va='center')
plt.title('Top-10 Brands by Number of Sales')
plt.xlabel('Number of Sales')
plt.ylabel('Brand')
plt.show()
```

This graph shows the top 10 car brands by number of sales. As you can see, the largest number of sold cars are Ford's cars

```python
max_price_by_brand = data.groupby('brand')['price'].max() #8th task
max_price_by_brand_sorted = max_price_by_brand.sort_values(ascending=False)
print(max_price_by_brand_sorted)
```

```python
the_most_exp_10 = max_price_by_brand_sorted.head(10)
the_most_exp_10_sorted = the_most_exp_10.sort_values(ascending=False)
print(the_most_exp_10)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=max_price_by_brand_sorted, y=max_price_by_brand_sorted.index, palette='viridis')
for index, value in enumerate(max_price_by_brand_sorted.values):
    plt.text(value, index, str(value), va='center')
plt.title('The most expensive car in every brand')
plt.xlabel('Price')
plt.ylabel('Brand')
plt.show()
```

This infographic is a pretty easy way to get the information you need. It shows the most expensive car sold in each brand in descending order.

```python
max_price_brand = data.groupby('brand')['price'].mean() #8th task
max_price_brand_sorted = max_price_brand.sort_values(ascending=False)
print(max_price_brand_sorted)
```

```python
the_most_exp_10_brands = max_price_brand_sorted.head(10)
the_most_exp_10_brands_sorted = the_most_exp_10_brands.sort_values(ascending=False)
print(the_most_exp_10_brands_sorted)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=the_most_exp_10_brands_sorted, y=the_most_exp_10_brands_sorted.index, palette='viridis')
for index, value in enumerate(the_most_exp_10_brands_sorted.values):
    plt.text(value, index, str(value), va='center')
plt.title('The most expensive brand')
plt.xlabel('Price')
plt.ylabel('Brand')
plt.show()
```

This infographic shows us the top 10 most expensive car brands sold at auction in North America. The data is arranged in descending order of price.

```python
plt.scatter(data['mileage'], data['price'], alpha=0.5)
plt.title('Dependence of car cost on mileage')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

This plot shows us the depence of car's price on mileage.

```python
filtered_data = data[(data['mileage'] <= 0.4e6) & (data['price'] <= 80000)]
plt.scatter(filtered_data['mileage'], filtered_data['price'], alpha=0.5)
plt.title('Dependence of car cost on mileage <= 0.4')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

This plot has the same information as previous one, but it filtered for better view of information.

```python
correlation_coefficient = data['mileage'].corr(data['price'])
print("Correlation coefficient of depence of mileage and cost:", correlation_coefficient)
```

This negative correlation coefficient shows that with an increase in the mileage of a car, its price decreases
